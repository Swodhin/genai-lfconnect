## 1. Variables & Dynamic Typing
A variable is a name bound to a value. Python figures out the type at runtime — you never declare it up front. This is called **dynamic typing**.

````python
model_name = "gpt-4"
temperature = 0.7
is_ready = True

print(model_name, type(model_name))
print(temperature, type(temperature))
print(is_ready, type(is_ready))

# Legal, but a readability smell: reassigning a name to a different type
temperature = "warming up"
print(temperature, type(temperature))
````

## 2. Core Data Types
The five types you'll reach for constantly: `int`, `float`, `str`, `bool`, and `None` (Python's "nothing here yet" value).

````python
token_count = 128          # int
avg_latency_sec = 1.42     # float
prompt = "Summarize this." # str
has_error = False          # bool
cached_response = None     # NoneType

for value in [token_count, avg_latency_sec, prompt, has_error, cached_response]:
    print(f"{value!r:30} -> {type(value)}")
````

## 3. Conditionals — if / elif / else
Indentation (4 spaces) defines a block in Python — there are no curly braces. `elif` checks another condition if the first was `False`; `else` catches everything remaining.

````python
score = 82

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"

print(f"Score {score} -> grade {grade}")
````

## 4. for Loops
`for` walks through any iterable: a string, a list, `range()`, a dict, and more. `enumerate()` hands you `(index, value)` pairs for free.

````python
prompts = ["Hi", "Explain RAG", "Bye"]

for i, prompt in enumerate(prompts):
    print(f"{i}: {prompt}")
````

## 5. while Loops
`while` repeats as long as its condition stays `True` — useful when you don't know the number of iterations in advance (like polling an API until a job finishes).

````python
attempts = 0

while attempts < 3:
    print(f"Attempt {attempts + 1}")
    attempts += 1

print("Done")
````

## 6. Lists & Tuples
Both are ordered and support indexing/slicing. Lists (`[]`) are **mutable** — they can grow, shrink, and change. Tuples (`()`) are **immutable** — once created, they can't be changed, which makes them a good fit for fixed groupings like coordinates.

````python
history = ["hi"]
history.append("how are you?")
print(history)

point = (10, 20)
print(point[0], point[1])

try:
    point[0] = 99
except TypeError as e:
    print(f"Can't do that: {e}")
````

## 7. Dicts & Sets
A `dict` maps keys to values — perfect for JSON-like configuration. A `set` holds only unique, unordered items — perfect for de-duplication and fast membership tests.

````python
config = {"model": "gpt-4", "temperature": 0.7}
print(config["model"])
config["max_tokens"] = 500
print(config)

seen = {"a", "b", "a"}
print(seen)
print("a" in seen)
````

## 8. Demo: Building a Tiny Word Counter
One `dict` plus one `for` loop is enough to build a small analytics tool. This exact pattern shows up again later when you tally tokens or count API responses by type.

````python
text = "the cat sat on the mat the cat ran"
counts = {}

for word in text.split():
    counts[word] = counts.get(word, 0) + 1

for word, n in sorted(counts.items()):
    print(f"{word}: {n}")
````
