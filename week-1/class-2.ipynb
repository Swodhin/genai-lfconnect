## 1. String Basics — Indexing, Slicing, Immutability
Strings are ordered sequences of characters. Index from 0; slice with `[start:stop]`. Strings are **immutable** — methods always return a *new* string rather than modifying the original.

````python
model = "gpt-4-turbo"
print(model[0])     # first character
print(model[-1])    # last character
print(model[0:3])   # slice
print(len(model))   # length
````

## 2. The String Method Toolbox
A handful of methods cover most day-to-day text cleanup: `.strip()`, `.lower()`/`.upper()`, `.split()`, `.join()`, `.replace()`, and `.startswith()`/`.endswith()`.

````python
raw = "  Hello, World!  "
print(repr(raw.strip()))
print(raw.strip().lower())

csv_line = "gpt-4,claude,llama"
models = csv_line.split(",")
print(models)
print(", ".join(models))

print("cat".replace("c", "b"))
print("sk-abc123".startswith("sk-"))
````

## 3. f-strings
Prefix a string with `f` and drop any expression inside `{}`. Format specifiers control decimals, commas, and percentages. f-strings are the standard way to build prompts dynamically.

````python
name = "Ada"
tokens_used = 1532
print(f"{name} used {tokens_used:,} tokens today")

accuracy = 0.9345
print(f"Accuracy: {accuracy:.1%}")

n = 3
topic = "retrieval-augmented generation"
prompt = f"Summarize {topic} in {n} sentences."
print(prompt)
````

## 4. What Is JSON?
JSON (JavaScript Object Notation) is plain text shaped like nested dicts and lists. Every LLM API sends and receives JSON. Type mapping: JSON `object` -> Python `dict`, JSON `array` -> Python `list`, and `true`/`false`/`null` map to `True`/`False`/`None`.

````python
example_json = '''
{
  "model": "gpt-4",
  "messages": [{"role": "user", "content": "Hi"}],
  "temperature": 0.7
}
'''
print(example_json)
````

## 5. Parsing JSON with `json.loads`
`json.loads(text)` turns a JSON string into a Python dict/list. Once parsed, it behaves like any normal dict — access with `["key"]`, loop with `.items()`.

````python
raw = '{"role": "assistant", "content": "Hello!"}'
data = json.loads(raw)
print(type(data))
print(data["content"])
````

## 6. Generating JSON with `json.dumps`
`json.dumps(obj)` turns a Python dict/list back into a JSON string. `indent=2` makes it human-readable; `sort_keys=True` gives consistent, diffable output.

````python
payload = {"model": "gpt-4", "temperature": 0.7, "stream": False}
print(json.dumps(payload, indent=2, sort_keys=True))
````

## 7. Reading & Writing Files
`open(path, mode)` opens a file — `"r"` read, `"w"` write, `"a"` append. Always use `with open(...) as f:` so the file closes automatically, even if an error occurs. `json.dump`/`json.load` work directly with file handles.

````python
with open("demo_log.json", "w") as f:
    json.dump({"status": "ok", "count": 3}, f, indent=2)

with open("demo_log.json") as f:
    loaded = json.load(f)

print(loaded)
````

## 8. Demo: From Raw Text to a Saved JSON Report
A small pipeline: parse a JSON "chat log" string, filter it with a list comprehension, and save a summary report to disk. This loop-parse-save pattern is exactly what logging pipelines do.

````python
log = '[{"role":"user","content":"hi"}, {"role":"assistant","content":"hello"}, {"role":"user","content":"how are you?"}]'

messages = json.loads(log)
user_msgs = [m["content"] for m in messages if m["role"] == "user"]

report = {"total_messages": len(messages), "user_messages": user_msgs}

with open("report.json", "w") as f:
    json.dump(report, f, indent=2)

with open("report.json") as f:
    print(f.read())
````

### Challenge 5 (Stretch) — Summarize Messages by Role
Using the `report.json`-style pattern from the demo, write code that takes a list of message dicts (make up at least 6, mixing `"user"` and `"assistant"` roles), counts how many messages belong to each role, and writes a summary dict to a file called `role_summary.json`.

**Acceptance criteria:** `role_summary.json` exists after running the cell and contains correct counts for each role.

````python
# TODO: build a message list, count messages per role, and write role_summary.json

messages = [
    {"role": "user", "content": "Hello"},
    {"role": "assistant", "content": "Hi there"},
    {"role": "user", "content": "How are you?"},
    {"role": "assistant", "content": "I am doing well"},
    {"role": "user", "content": "Tell me about Python"},
    {"role": "assistant", "content": "Python is a programming language"}
]

role_summary = {}

for message in messages:
    role = message["role"]
    role_summary[role] = role_summary.get(role, 0) + 1

with open("role_summary.json", "w") as file:
    json.dump(role_summary, file, indent=2, sort_keys=True)

with open("role_summary.json", "r") as file:
    print(file.read())
````
